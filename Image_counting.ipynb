```python
#Import the necesary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load image and convert it to a gray scale
image = cv2.imread('coins.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray');
```

```python
#to reduce capturing noise we use the gausianBlur() to detect the edges
blur = cv2.GaussianBlur(gray, (11,11), 0)
plt.imshow(blur, cmap='gray')
```

```python
#Canny Edge detector or the edges
canny = cv2.Canny(blur, 30, 150, 3)
plt.imshow(canny, cmap='gray')
```

```python
#have the edges well distinguishable
dilated = cv2.dilate(canny, (1,1), iterations = 2)
plt.imshow(dilated, cmap='gray')

#you can see edges are well clear now
```

```python
#find contours in the image
(cnt, heirarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0,255,0), 2)

plt.imshow(rgb)
```

```python
#find the number of coins from the image file
print('Coins in the image: ', len(cnt))
plt.show()
```
